## Data Frame oluşturma


```python
df = pd.DataFrame(anime_data)
```

##Arayüz

```python
genre_dropdown = widgets.Dropdown(
    options=df["Genre"].unique(),
    description="Tür:"
)

button = widgets.Button(description="🎯 Öner")
surprise_button = widgets.Button(description="🎲 Surprise Me")
reset_button = widgets.Button(description="🔄 Reset")

output = widgets.Output()
```

##Öneri

```python
def recommend(b):
    with output:
        clear_output()

        result = df[df["Genre"] == genre_dropdown.value]

        if result.empty:
            print("Öneri bulunamadı 😢")
        else:
            for _, row in result.iterrows():
                print("Anime:", row["Anime"])
                print("Tür:", row["Genre"])
                print("Tip:", row["Type"])
                print("Puan:", row["Score"])

                display(Image(url=row["Image"], width=200))
                print("-" * 40)

button.on_click(recommend)
```

##Beni Şaşırt


```python
def surprise_me(b):
    with output:
        clear_output()

        row = df.sample(1).iloc[0]

        print("🎉 Surprise Anime")
        print("Anime:", row["Anime"])
        print("Tür:", row["Genre"])
        print("Tip:", row["Type"])
        print("Puan:", row["Score"])

        display(Image(url=row["Image"], width=200))

surprise_button.on_click(surprise_me)
```

##Reset Butonu


```python
def reset(b):
    with output:
        clear_output()

    genre_dropdown.value = df["Genre"].unique()[0]

reset_button.on_click(reset)
```

```python
display(
    genre_dropdown,
    button,
    surprise_button,
    reset_button,
    output
)
```
